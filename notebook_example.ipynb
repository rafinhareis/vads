```python
import vads
import matplotlib.pyplot as plt
import numpy as np
```

```python
#open file .xy it will create a file .ARPES to open later with better computational cost
spec =  vads.load_spectrum('Examples/Bi2Se3_21eV.xy')
#spec =  vads.load_spectrum('Examples/Bi2Se3_21eV.arpes') # file .arpes saved
```

```python
vads.SBZ() #shown de brillouin zone and the values of the simmetry points
```

```python
spec2D = vads.getSlice(spec,axis = 2,axisValue= -2,sliceIntegration=999) #get a slice 2D for the spectrum 3D
spec2D = vads.kwarp(spec2D,hv = 21) #apply the k-warping
spec2D = vads.interpolation(spec2D,filter_sv = True) #interpolate the data to better visualization
fig,ax=plt.subplots(figsize=(10,8))
vads.quickPlot(spec2D,axis = ax) #plot 2D data
#ax.set_ylim(5,0)
ax.set_xlim(-0.4,0.4)
        
```

```python
vads.secondDerivativeExplorer(spec2D,cmap = 'bone_r') #visualization first adn second derivative
```

```python
vads.explorer(spec) #interative visualization for 3D data
```

```python
vads.explorer(vads.kwarp(spec,hv=21)) #interative visualization for 3D data with k-warping applied
```
